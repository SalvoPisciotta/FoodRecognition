# Training history

```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def plot_history(history, keys, title="Training history", labels=("epochs", "metrics"), figsize=(10,5)):
    
    fig = plt.figure(figsize=figsize)
    
    # take all pairs of training and val metrics
    for i in range(0, len(keys), 2):
        m, val_m = keys[i], keys[i+1]
        
        plt.plot(history[m])
        plt.plot(history[val_m])
        
    plt.xlabel(labels[0])
    plt.ylabel(labels[1])

    # Adding legend
    plt.legend(keys,
          title ="Legend",
          loc ="upper left",
          bbox_to_anchor =(0.8, 0, 0.5, 1))
    plt.title(title)
    plt.grid(True)
    plt.show()
```

### Model A

```python
MODEL_PATH = "../models/model_A/"
df_a = pandas.read_csv(MODEL_PATH+"train_history_model_A.csv")
```

```python
plot_history(df_a,['loss','val_loss', 'iou_score','val_iou_score'], "Model A", figsize=(9,6))
```
